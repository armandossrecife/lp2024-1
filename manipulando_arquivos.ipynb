## Importância dos arquivos de dados e programas executáveis

- Um **arquivo de computado**r é uma coleção de informações digitalizadas armazenadas em um dispositivo de armazenamento, como um disco rígido, unidade flash USB ou nuvem (computador com acesso a Internet).

- Existem **arquivos de texto** e **arquivos binários**, por exemplo arquivos que armazenam documentos de texto, imagens, música, vídeo, dados, configurações e executáveis.

- Os arquivos possuem um **nome** e uma **extensão**, que indica o tipo de arquivo e o software (programa) necessário para abri-lo

## Qual a função dos arquivos digitais?

- Arquivos de computador servem como meios de armazenamento e compartilhamento de informação em um sistema de computador.

- Eles são utilizados para armazenar dados em modo texto e modo binário.

- Os arquivos são usados para preservar informações e permitir que sejam acessadas, compartilhadas e modificadas de maneira fácil e conveniente

- Eles são a base para armazenamento e compartilhamento de informações em um computador, e são utilizados em praticamente todos os aspectos da informática e da tecnologia da informação

## Qual a diferença entre **Arquivos texto** e **Arquivos binários**?

- **Arquivos de texto** são arquivos que contêm texto legível por humanos.

- Eles são codificados usando codificações como ASCII ou UTF-8 (representação de texto em idiomas que tenham caracteres especiais como Chines, Hebraico, Espanhol, Russo, ...).

- Eles podem ser lidos e editados com editores de texto, funcionar para troca de informações (arquivos .csv, .json), exibição de informações na Web (por exemplo: .html), ou armazenar instruções de programas de computador (arquivos de código-fonte: .py, .c, .java, .js)

- **Arquivos binários** são arquivos que contêm dados binários, que são dados codificados em formato binário (0s e 1s).

- Os arquivos binários são usados para armazenar informações em formato não legível para humanos, como imagens, áudios, vídeos, fontes, executáveis, etc.

- Os arquivos binários geralmente precisam de aplicativos específicos para serem lidos e editados.

## File System (Sistema de Arquivos) de um Sistema Operacional

- A forma como um computador organiza, nomeia, armazena e manipula arquivos é globalmente chamada de sistema de arquivos.

- A maioria dos computadores tem pelo menos um sistema de arquivos. O controle do sistema de arquivos depende do sistema operacional que roda no computador

- Windows: FAT16, FAT32, NTFS e exFAT, e etc.
- Linux e Android: ext2, ext3, ext4, e etc.
- MacOS e IOS: HFS, APFS

## Fluxo básico de leitura e exibição dos dados de um arquivo local

- O fluxo de dados de leitura de um arquivo envolve vários componentes do sistema, incluindo o **processador**, a **memória RAM**, o **disco rígido**, o **barramento de dados** e o **display**.

- O processador solicita ao sistema operacional que leia um arquivo específico armazenado no disco rígido.

- O sistema operacional faz a requisição de leitura ao controlador de disco rígido, que lê os dados do arquivo em questão e os transfere para a memória RAM.

- O processador acessa os dados no armazenamento na memória RAM para iniciar o processamento dos dados.

- O processador usa o barramento de dados para enviar os dados processados para o controlador de vídeo.

- O controlador de vídeo envia os dados para o display , onde eles são exibidos na tela.

### Lendo o conteúdo de um arquivo texto existente

função open(...)
https://docs.python.org/3/library/functions.html#open

````python
meu_arquivo = open('nomes.txt', mode='r')
conteudo = meu_arquivo.read()
print(conteudo)
meu_arquivo.close()
````

### Lendo um arquivo texto linha a linha

Obs: lembre-se do caractere especial '\n' que significa uma nova linha

````python
meu_arquivo = open('nomes.txt', mode='r')
for linha in meu_arquivo:
  print(linha)
meu_arquivo.close()
````

````python
meu_arquivo = open('nomes.txt', mode='r')
for linha in meu_arquivo:
  print(linha, end='')
meu_arquivo.close()
````

### Contando a quantidade de caracteres de um arquivo texto

````python
meu_arquivo = open('nomes.txt', mode='r')
conteudo = meu_arquivo.read()
qtd_caracteres = len(conteudo)
print(f'Quantidade de caracteres do arquivo nomes.txt: {qtd_caracteres}')
````

### Contando a quantidade de linhas de um arquivo texto

````python
meu_arquivo = open('nomes.txt', mode='r')
qtd_linhas = 0
for linha in meu_arquivo:
  qtd_linhas += 1
meu_arquivo.close()
print(f'Quantidade de linhas do arquivo nomes.txt: {qtd_linhas}')
````

### Usando o comando with para simplificar o gerenciamento de recursos

A sintaxe **with** elimina a necessidade de fechar explicitamente o arquivo, evitando erros e problemas de memória.

```python
with open(filename, mode):
  comand1
  comand2
  ..
  comandN
```


````python
with open('nomes.txt', mode='r') as meu_arquivo:
  conteudo = meu_arquivo.read()
  print(conteudo)
````

### Métodos de acesso a um arquivo usando Python

- "r": Leitura (read) - Valor padrão. Abre um arquivo para leitura, levanta um erro se o arquivo não existir.
- "a": Anexar (append) - Abre um arquivo para anexar conteúdo, cria o arquivo se ele não existir.
- "w": Write (append) - Abre um arquivo para escrita, cria o arquivo se ele não existir.

### Sobrescrevendo o conteúdo de um arquivo texto

````python
with open('nomes.txt', mode='w') as meu_arquivo:
  novo_conteudo = 'Armando Soares Sousa'
  meu_arquivo.write(novo_conteudo)
````

````python
with open('nomes.txt', mode='r') as meu_arquivo:
  conteudo = meu_arquivo.read()
  print(conteudo)
````

### Especificando como o arquivo deve ser tratado, como texto ou como binário

- "t": Texto - Valor padrão. modo de texto
- "b": Binário - Modo binário (por exemplo, imagens, áudio, vídeo)

**Modo Texto ("t")**:

- Interpretação: Os caracteres do arquivo são interpretados como caracteres Unicode.
- Codificação: O arquivo é codificado usando a codificação padrão do sistema ou uma codificação específica (se definida).
- Novas linhas: As novas linhas no arquivo são representadas pelo caractere especial \n (nova linha).

Exemplo: Ideal para arquivos que contêm texto puro, como scripts Python, arquivos de configuração, etc.

**Modo Binário ("b")**:

- Interpretação: Os bytes do arquivo são interpretados como valores brutos.
Codificação: Nenhuma codificação é aplicada. Os bytes do arquivo são lidos e gravados diretamente.
- Novas linhas: As novas linhas no arquivo podem ser representadas por diferentes combinações de bytes, dependendo do sistema operacional e do formato do arquivo.
- Exemplo: Ideal para arquivos que contêm dados binários, como imagens, arquivos de música, vídeos, etc.
